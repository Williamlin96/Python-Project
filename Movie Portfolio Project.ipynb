```python
# Import libraries

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plots we will create
```

```python
# Read the data

df = pd.read_csv(r'C:\Users\USER\Downloads\Documents\movies.csv')
```

```python
# Look the overview of the data

df.head()
```

## Data Cleaning

## Check missing and replacing values

```python
# Check if there any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Replacing missing values

cols_with_missing = ["budget", "gross"]

df[cols_with_missing].hist(bins=30)
plt.show()

df = df.dropna()
```

## Correct data types and column

```python
# Data types for our columns

df.dtypes
```

```python
# Change data types from float to int

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Change the format to year,month,date


df['yearcorrect'] = pd.to_datetime(df['released'].str.extract(r'(\w+ \d+, \d+)', expand=False), format='%B %d, %Y')
```

```python
df
```

```python
df.dtypes
```

```python
# Create correct year column

df['yearcorrect'] = df['yearcorrect'].astype(str).str[:4]
```

```python
df
```

## Sorting Columns

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

## Removing duplicates

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

## Finding correlation in the data

```python
df['budget'].head(10).sort_values(ascending=False)
```

```python
# Scatter plot with budget and gross

plt.scatter(x='budget', y='gross', data=df, alpha=0.5)

plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.title('Budget vs Gross Earnings')

plt.show()
```

```python
# Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color" : "red"}, line_kws={"color" : "blue"})
```

## Start looking at correlation

```python
df.corr() #pearson(default), kendall, spearman

# However, this correlation only working on numeric field
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.title('Correlation Matrix for Numeric Features')
```

```python
# Looks at object columns

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
         df_numerized[col_name] = df_numerized[col_name].astype('category')
         df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized          
```

```python
correlation_matrix = df_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.xlabel('Movies Features')
plt.ylabel('Movies Features')
plt.title('Correlation Matrix for Numeric Features')
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
